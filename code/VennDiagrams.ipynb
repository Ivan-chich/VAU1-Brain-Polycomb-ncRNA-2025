# Venn diagrams

```python
# import libraries:
import matplotlib.pyplot as plt
from matplotlib_venn import venn2
import matplotlib.patches as mpatches
```

```python
# read files:
with open('../data/tf_list_Alina.txt', 'r') as file1:
    list1 = list(map(lambda x: x.strip(), file1.readlines()))

with open('../data/GO_TF_set.txt', 'r') as file2:
    list2 = list(map(lambda x: x.strip(), file2.readlines()))
```

```python
# Преобразуем списки в множества:
set1 = set(list1)
set2 = set(list2)
```

```python
print('Alina TF list:', len(set1))
print('Ivan TF list:', len(set2))
```

```python
# Объединение
union = set1.union(set2)  # или set1 | set2
print("Объединение:", len(union))
```

```python
# Пересечение
intersection = set2.intersection(set1)  # или set1 & set2
print("Пересечение:", len(intersection))
```

```python
# Подготовка данных для диаграммы Венна
only_set1 = set1 - intersection  # Элементы только в первом множестве
only_set2 = set2 - intersection  # Элементы только во втором множестве
```

```python
# Функция для форматирования списка с учетом длины
def format_list(elements):
    sorted_elements = sorted(elements)
    if len(sorted_elements) <= 10:
        return '\n'.join(sorted_elements) + f'\n({len(sorted_elements)})'
    else:
        return '\n'.join(sorted_elements[:10]) + f'\n... ({len(sorted_elements)})'
```

```python
# Создание диаграммы Венна с заданием цветов
plt.figure(figsize=(8, 6))
venn_diagram = venn2(subsets=(len(only_set1), len(only_set2), len(intersection)),
                     set_labels=('', ''),
                     alpha=0.5)  # Устанавливаем прозрачность

# Задаем новые цвета для каждого множества
venn_diagram.get_patch_by_id('10').set_color('#FF9999')  # Насыщенный красный для Alina TF List
venn_diagram.get_patch_by_id('01').set_color('#99CCFF')  # Насыщенный голубой для Ivan TF List
venn_diagram.get_patch_by_id('11').set_color('#CCCCCC')   # Серый для пересечения

# Подпись элементов на диаграмме без заголовков
venn_diagram.get_label_by_id('10').set_text(format_list(only_set1))  # Элементы только в первом множестве
venn_diagram.get_label_by_id('01').set_text(format_list(only_set2))  # Элементы только во втором множестве
venn_diagram.get_label_by_id('11').set_text(format_list(intersection))  # Элементы пересечения

# Добавление легенды с квадратиками соответствующего цвета
legend_elements = [
    mpatches.Patch(color='#FF9999', label='Alina TF List'),
    mpatches.Patch(color='#99CCFF', label='Ivan TF List'),
    mpatches.Patch(color='#CCCCCC', label='Пересечение')
]

plt.legend(handles=legend_elements, loc='upper right')

# Отображение диаграммы без заголовков кружочков
plt.title("Диаграмма Венна для Alina TF List и Ivan TF List")
plt.show()
```

### PDX1 (Pancreatic and Duodenal Homeobox 1):
Функция: PDX1 является транскрипционным фактором, который играет ключевую роль в развитии поджелудочной железы и в регуляции функции бета-клеток, отвечающих за выработку инсулина. Он необходим для формирования поджелудочной железы в эмбриональном развитии и поддерживает нормальную секрецию инсулина у взрослых.

### PRDM6 (PR Domain Containing 6):
Функция: PRDM6 является транскрипционным фактором, который участвует в регуляции клеточной дифференцировки и развития. Он может играть роль в поддержании стволовых клеток и в процессе ангиогенеза (образования новых кровеносных сосудов). Также PRDM6 может быть вовлечен в патогенез некоторых заболеваний, включая рак.

### TTF1 (Thyroid Transcription Factor 1):
Функция: TTF1, также известный как NKX2-1, является транскрипционным фактором, который регулирует экспрессию генов, связанных с развитием щитовидной железы, легких и мозга. Он играет важную роль в формировании легких и щитовидной железы во время эмбрионального развития и поддерживает нормальную функцию этих органов у взрослых.

### ZNF503 (Zinc Finger Protein 503):
Функция: ZNF503 относится к семейству белков с цинковыми пальцами и может действовать как транскрипционный фактор. Хотя его функции менее изучены по сравнению с другими генами из этого списка, предполагается, что он участвует в регуляции экспрессии генов и может играть роль в клеточной дифференцировке и развитии.

### ALX1 (ALX Homeobox 1):
Функция: ALX1 является транскрипционным фактором, который играет важную роль в развитии лицевых структур и формировании черепа. Он участвует в регуляции клеточной дифференцировки и может быть вовлечен в процессы, связанные с развитием зубов и других тканей.

### ALX4 (ALX Homeobox 4):
Функция: ALX4 также является транскрипционным фактором и участвует в развитии лицевых структур и конечностей. Он играет роль в регуляции клеточной миграции и дифференцировки во время эмбрионального развития.

### ARX (Aristaless Related Homeobox):
Функция: ARX является транскрипционным фактором, который играет ключевую роль в развитии нервной системы. Он участвует в нейрогенезе и может быть вовлечен в регуляцию экспрессии генов, связанных с развитием нейронов и глиальных клеток. Мутации в этом гене связаны с различными неврологическими расстройствами.

### BARHL2 (BarH-like Homeobox 2):
Функция: BARHL2 является транскрипционным фактором, который участвует в развитии нервной системы и формировании органов чувств. Он может играть роль в регуляции нейрогенеза и дифференцировки нейронов.

### BHLHE22 (Basic Helix-Loop-Helix Family Member E22):
Функция: BHLHE22 является членом семейства белков с базовой спирально-цилиндрической структурой (bHLH) и может действовать как транскрипционный фактор. Он участвует в регуляции биологических ритмов и может играть роль в процессе адаптации к изменениям окружающей среды.

### BSX (Brain and Spinal Cord Homeobox):
Функция: BSX является транскрипционным фактором, который играет важную роль в развитии мозга и спинного мозга. Он участвует в регуляции экспрессии генов, связанных с развитием нейронов и формированием нервной системы.

### CBX2 (Chromobox 2):
Функция: CBX2 является членом семейства белков с хромодоменами и может действовать как транскрипционный репрессор. Он участвует в регуляции генной экспрессии во время эмбрионального развития и может играть роль в поддержании стволовых клеток.

### DBX1 (Developing Brain Homeobox 1):
Функция: DBX1 является транскрипционным фактором, который играет важную роль в развитии спинного мозга и формировании нейронов. Он участвует в нейрогенезе и может влиять на развитие различных типов нейронов.

### DMBX1 (Dmbx1 Homolog):
Функция: DMBX1 является транскрипционным фактором, который участвует в развитии мозга и формировании нервной системы. Он может играть роль в регуляции экспрессии генов, связанных с развитием нейронов.

### DMRT3 (Doublesex and Mab-3 Related Transcription Factor 3):
Функция: DMRT3 является транскрипционным фактором, который играет важную роль в половой дифференциации и развитии половых органов у позвоночных животных. Он также участвует в регуляции нейрогенеза и может влиять на поведение животных.
