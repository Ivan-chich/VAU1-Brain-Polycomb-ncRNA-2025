## Polycomb gene-target extraction

```python
import pandas as pd
```

```python
#экстаркция данных о генах-таргетов поликомба
table = pd.read_csv('../data/polycomb_dots_hand_coords_update.tsv', sep = '\t')
```

```python
table.info
```

```python
table.head(10)
```

```python
gene_list1 = table['gene1']
gene_list2 = table['gene2']
gene_list = gene_list1 + gene_list2
```

```python
final_gene_list = []
for gene in gene_list:
    if ',' in gene:
        x = gene.split(',')
        final_gene_list.extend(x)
    else: final_gene_list.append(gene)
        
```

```python
len(final_gene_list)
```

```python
#итоговый список генов 
final_gene_set = set(final_gene_list)
len(final_gene_set)
```

```python
final_gene_set
```

```python
#запись генов в файл
with open('gene_list.txt', 'w') as file:
    file.writelines(f"{item}\n" for item in final_gene_set)
```

```python
#поиск генов транскрипционных факторов по https://simchrom.intbio.org/
transcription_factors = pd.read_csv('../data/simchrom_intbio_TF.csv', sep = ',')
```

```python
len(transcription_factors)
```

```python
tf_list = []
for gene in transcription_factors['Gene']:
    gene_name = gene.replace('undefined','')
    tf_list.append(gene_name)
```

```python
len(set(tf_list))
```

```python
#отбор транскрипционных факторов 
selected_tf_from_intbio = [name for name in final_gene_set if name in tf_list]
len(selected_tf_from_intbio)
```

## TF selection from database https://humantfs.ccbr.utoronto.ca/allTFs.php

```python
#оставляем только транскрипционные факторы
TF_data = pd.read_csv('../data/humantfs_TF.csv')
TF_data_filtered = TF_data[TF_data['Is TF?'] == 'Yes']
```

```python
#создаем список генов
tf_list_new = []
for gene in TF_data_filtered['HGNC symbol']:
    tf_list_new.append(gene)
```

```python
tf_list_new = list(set(tf_list_new))
len((tf_list_new))
```

```python
tf_set_new = set(tf_list_new)
len(tf_set_new)
```

```python
selected_tf_from_toronto = [name for name in final_gene_set if name in tf_set_new]
len(selected_tf_from_toronto)
```

```python
with open('tf_list.txt', 'w') as file:
    file.writelines(f"{item}\n" for item in selected_tf_from_toronto)
```

## Initial table sorting by selected TF

```python
#df['Match'] = df[['Gene1', 'Gene2']].isin(gene_set).any(axis=1)
table['match_tf'] = table[['gene1', 'gene2']].isin(selected_tf_from_toronto).any(axis = 1)
```

```python
def check_gene_match(table):
    genes1 = set(table['gene1'].split(','))  
    genes2 = set(table['gene2'].split(',')) 
    return not genes1.isdisjoint(selected_tf_from_intbio) or not genes2.isdisjoint(selected_tf_from_intbio)


table['Match'] = table.apply(check_gene_match, axis=1)
```

```python
len(table[table['Match'] == True])
```

## GO enrichment for selected TF taregts

```python
from gprofiler import GProfiler
```

```python
gp = GProfiler(return_dataframe=True)

results = gp.profile(organism='hsapiens', query= list(set(targets)))

go_results = results[results['source'].str.contains('GO')]

print(go_results[['name', 'p_value', 'source']])
```

```python
go_results.to_csv('go_results.txt', index=False)
```

```python
go_results.columns
```

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
def plot_go_terms(df):
    df = df.sort_values("p_value").head(30)  # Берём топ-10 значений по p-value
#    plt.figure(figsize=(10, 5))
    sns.barplot(y=df["name"], x=-df["p_value"].apply(lambda x: np.log10(x)), palette="viridis")
    plt.xlabel("-log10(p-value)")
    plt.ylabel("GO Term")
    plt.show()

plot_go_terms(go_results)
```
